```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ind_hotels=pd.read_csv("./india_all_restaurants_details.csv")
ind_hotels.head()
```

```python
ind_hotels.info()
```

```python
#Deleting Unnnecessary Columns
ind_hotels=ind_hotels.drop(['zomato_url','famous_food','telephone','Unnamed: 0','sno'],axis=1) 
```

```python

#Removing the Duplicates
ind_hotels.duplicated().sum()
ind_hotels.drop_duplicates(inplace=True)

#Remove the NaN values from the dataset
ind_hotels.isnull().sum()
ind_hotels.dropna(how='any',inplace=True)

```

```python
ind_hotels.head()
```

```python

ind_hotels = ind_hotels.loc[ind_hotels.area !='NaN'].reset_index(drop=True)
ind_hotels['area'].unique()
```

```python
#renaming the column ‘cost_for_two’ as ‘cost’.
ind_hotels = ind_hotels.rename(columns={'cost_for_two':'cost'}) 
ind_hotels.columns
```

```python
#checking the unique values in the columns
ind_hotels['city'].unique()
```

```python
ind_hotels['rating'].unique()
```

```python
ind_hotels = ind_hotels.loc[ind_hotels.rating !='NEW']
ind_hotels = ind_hotels.loc[ind_hotels.area !='NaN'].reset_index(drop=True)

ind_hotels = ind_hotels.loc[ind_hotels.rating !='Nové'].reset_index(drop=True)
ind_hotels['rating'] = ind_hotels['rating'].astype(float)
ind_hotels['rating'].unique()
```

```python
#Removing commas, full stops in df[‘Cost’] column also converting  into a data type float.
ind_hotels['cost'] = ind_hotels['cost'].astype(str) #Converting the cost to string
ind_hotels['cost'] = ind_hotels['cost'].apply(lambda i: i.replace(',','.')) 
ind_hotels['cost'] = ind_hotels['cost'].astype(float) # Changing the cost from object string to Float as cost should be in float

ind_hotels['price_range'] = ind_hotels['price_range'].astype(int) # Changing from float to int
ind_hotels['rating_count'] = ind_hotels['rating_count'].astype(int) # Changing from float to int

ind_hotels.info()
```

```python
ind_hotels.isna().sum()
```

```python
#Converting ‘name’ column to camel-case to standardize .
ind_hotels.name = ind_hotels.name.apply(lambda j:j.title())
ind_hotels.name
```

```python
ind_hotels.online_order.replace((True,False),('Yes','No'),inplace=True)
ind_hotels['online_order']
```

```python
ind_hotels.delivery_only.replace((True,False),('Yes','No'),inplace=True)
ind_hotels['delivery_only']
```

```python
ind_hotels.table_reservation.replace((True,False),('Yes','No'),inplace=True)
ind_hotels['table_reservation']
```

```python
def Recommend(city,rating_L,rating_H,cost_max):
    df_filter = pd.DataFrame(columns=['name', 'city', 'area', 'rating','rating_count','cusine','cost','address','timings','online_order','table_reservation','delivery_only'])
    sorted_df = pd.DataFrame(columns=['name', 'city', 'area', 'rating','rating_count','cusine','cost','address','timings','online_order','table_reservation','delivery_only'])

    for i in range(1,len(ind_hotels)):    
        if(ind_hotels.rating[i]>rating_L and ind_hotels.rating[i]<rating_H):
            if(ind_hotels.city[i]==city):
                if(ind_hotels.cost[i]>199 and ind_hotels.cost[i]<cost_max):
                    restaurants=ind_hotels.iloc[i]
                    restaurants=dict(restaurants)
                    df_filter = df_filter.append(restaurants, ignore_index = True)

    sorted_df=(df_filter.sort_values(["rating"],ascending = False))
    return(sorted_df) 
city_input=input("Please enter your city:")
high_lim=float(input("please Enter higher limit of rating: "))
low_lim=float(input("please Enter lower limit of rating: "))
cost_lim=int(input("Please Enter maximum cost for two: "))
df3=Recommend(city_input,low_lim,high_lim,cost_lim)
limit_rcd=int(input("Top N restaurants :"))
df3.head(limit_rcd)
```
